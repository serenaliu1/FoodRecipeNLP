```python
cv = KFold(n_splits=10, random_state=42, shuffle=True)

svc = SVC()
svc.fit(X_fdr, y)
y_pred = svc.predict(X_fdr)
scores = cross_val_score(svc, X_fdr, y, scoring='accuracy', cv=cv, n_jobs=-1)
print('10-Fold accuracy of SVC : %.3f (std = %.3f)' % (mean(scores), std(scores)))
print('F1 score of SVC : %.3f' % (mean(f1_score(y, y_pred, average='weighted'))))
accuracy_score(y, y_pred)
```
